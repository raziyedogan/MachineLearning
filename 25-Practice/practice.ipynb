Logistic regression metodu ile ve grid search cross validation ile elde ettiğimiz en iyi parametreleri kullanarak yeni bir model geliştirelim ve bunu test datası ile test edelim.

```python
from sklearn.datasets import load_iris

import pandas as pd
import numpy as np
```

```python
iris = load_iris()
```

```python
x = iris.data  #feature'ler
y = iris.target  #class
```

```python
x = x[:100,:]  #İlk 100 değer 2 tane class'ı içerir.  
y = y[:100]
```

```python
x = (x-np.min(x))/(np.max(x)-np.min(x))
```

```python
#train test split

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV

grid = {"C":np.logspace(-3,3,7),"penalty":["l1","l2"]} 

logreg = LogisticRegression()

logreg_cv = GridSearchCV(logreg, grid, cv=10)
logreg_cv.fit(x_train,y_train)
```

```python
print("tuned hyperparameters: (best parameters): ",logreg_cv.best_params_)
```

```python
print("accuracy: ",logreg_cv.best_score_)
```

En uygun C değerinin 1 olduğunu yukarıda tespit ettik. Dolayısıyla C=1 ile algoritmayı çalıştıralım.

```python
logreg2 = LogisticRegression(C=1, penalty="l2")
logreg2.fit(x_train,y_train)
print("score: ",logreg2.score(x_test,y_test))
```
