## Model Selection

Model selection, en iyi parametreleri seçmektir. Parametreleri şu şekilde ayırabiliriz:

    1- Öğrenilen parametreler: Bu parametreleri model kendi öğrenecektir dolayısıyla bu parametreleri bizim seçmemiz söz konusu değildir.
    
    2- Seçilen parametreler: Bunlara hyperparameter denir. Örneğin KNN algoritmasında k değeri.
    
Mesela bir model oluşturduk ve birde test datası oluşturduk. Accuracy değeri %89 çıktı diyelim. Sonra aynı modeli başka bir test datası ile tekrar denedik ve accuracy değeri %60 çıktı.  %89 ile %60 arasında ciddi bir fark vardır. Dolayısıyla bu problemi ortadan kaldırabilmemiz için K-Fold Cross Validation yönteminden faydalanabiliriz. Modeli eğitmek için K-Fold Cross Validation yöntemin yöntemini kullanacağız.

## K-Fold Cross Validation

K-Fold Cross Validation: avoid overfitting.

Overfitting, bir data'yı ezberlemek demektir. Eğer model bir data'yı ezberlerse başka bir test data'sı verildiğinde doğru tahminler gerçekleştirilemez. Çünkü model, data'yı ezberlemiştir, öğrenmemiştir. Dolayısıyla overfitting durumunu önlemek gerekmektedir. Bunu önleme yöntemlerinden birisi K-Fold Cross Validation 'dır.

<br><br>

Data'yı test ve train setlerine ayırdık. Train setine ile model eğitilir. Test seti ile model test edilir. Sonuç olarak accuracy değeri %80 diyelim. Tek bir deneme yapılmıştır fakat bir modelden emin olabilmek için o modeli birden fazla kez denemeliyiz. Yani tekrardan train ve test setlerine ayırıp, tekrardan accuracy değerine bakmalıyız. Bu şekilde birden fazla deneme yapmalıyız. Bunu k-fold cross validation yöntemi ile yapabiliriz.

K-fold cross validation yöntemi şudur:

Data'yı train ve test setlerine böleriz. Sonrasında test setini bir kenara alıyoruz ve hiçbir şekilde modeli eğitirken veya test ederken test setini başlangıçta kullanmayacağız. En son kullanacağız.

Yukarıdaki görselde k-fold cross validation işlemi ifade edilmiştir. Açıklayalım.

Train setini k kez böleriz. Mesela k=3 olsun. Data 3 eşit parçaya bölünür. İlk bölme işleminde mavi kısımlar train değerleri olur. Diğer parçada validation olur. Train parçaları birleşiktir yani tek bir parçadırlar ve bunlarla eğitilir, validation parçası ile de test yapılır. Test data'sı bir kenarda durmaya devam etmekte ve test datası ile henüz test işlemi yapılmamıştır. Validation parçası ile test işlemi sağlanmıştır.

Sonrasında tekrar parçalara bölüyoruz fakat bu sefer train ve validation parçaları farklıdır. Bu sefer data'nın bir önceki bölme işleminden farklı olarak belirlenmiş Train parçaları ile eğitim yapılır ve validation parçası ile test işlemi yapılır yani doğrulama yapılır. 

Sonrasında bir öncekilerden farklı şekilde bölme işlemi gerçekleştirilir. Train parçaları tek bir settir, Yani bu iki train parçalarının toplamı ile eğitim yapılır. Validation parçası ile test yapılır.

Bu işlemler sayesinde 3 farklı accuracy değeri elde edilir. Sonrasında elde edilen tüm accuracy değerlerinin ortalaması alınır ve tek bir tane accuracy değeri elde edilir.

Böylelikle data'yı train ve validation setlere göre eğittik ve test ettik. Öncesinde kenara ayırdığımız test seti ile yukarıdaki işlemler sonucunda oluşan modeli en son test ediyoruz. Bu test setini model daha öncesinde hiç görmediği için gerçek hayatta çıkabilecek sorunların çözümünde yardımcı oluyor.

k değeri train setini kaça böleceğimizi belirtir. Train seti k defa bölünür ve her bölme işleminde train ve validation setleri farklı sample'lerden oluşur. Sonrasında her bölmeden bir accuracy değeri elde edilir. Elde edilen tüm accuracy değerlerin ortalaması alınarak tek bir accuracy değeri elde edilir.

Modelin Train data'yı ezberlemesini önlemek için yani overfitting durumunu engellemek için k-fold cross validation yöntemini kullanırız. 

```python
from sklearn.datasets import load_iris

import pandas as pd
import numpy as np
```

```python
iris = load_iris()
```

```python
"""
#iris veri seti içerisinde numpy array'leri barındıran bir yapıdır. Bir data frame oluşturarak veri setini daha anlaşılır 
#bir şekilde görelim.

data = iris.data
feature_names = iris.feature_names
y = iris.target  # target feature'sinde 0,1,2 şeklinde 3 tane sınıf vardır.

df = pd.DataFrame(data, columns=feature_names)

df["sinif"] = y

df.head()
"""
```

```python
x = iris.data  #feature'ler
y = iris.target  #class
```

```python
#normalizasyon yaparız çünkü KNN algoritması kullanacağız. 

x = (x-np.min(x))/(np.max(x)-np.min(x))
```

```python
#train test split

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3)
```

```python
#KNN Model

from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=3)
```

```python
# K fold cross validation

from sklearn.model_selection import cross_val_score

accuracies = cross_val_score(estimator=knn, X=x_train, y=y_train, cv=10)
```

estimator=knn parametresi ile cross validation yapılırken knn algoritmasının kullanılması sağlanır. k-fold cross validation'daki k değeri cv=10 parametresi ile 10 olarak belirlenmiştir.

```python
accuracies
```

k-fold cross validation sonucunda elde edilen 10 tane accuracy değerini yukarıdaki çıktıda görüyoruz.

Şimdi elde edilen accuracy değerlerin ortalamasını alarak tek bir accuracy değeri elde edelim.

```python
print("average accuracy: ",np.mean(accuracies))
print("average std: ",np.std(accuracies))
```

np.std(accuracies) ile standard deviation hesaplanmıştır. Standard deviation, data'nın sonucunda elde edilen accuracy'in yayılımıdır. Yani ne kadar tutarlı olduğu denilebilir. 

Ortalama accuracy değeri %97 ve standard deviation değeri 0.04 'tür. Yani data tutarlı denilebilir.

Modelin tutarlı olduğuna karar verdikten sonra ve %97 accuracy değerinden KNN algoritmasındaki k=3 değerinin iyi bir değer olduğuna karar verdikten sonra ilk başta oluşturduğumuz test değerleri (x_test,y_test) ile test işlemi yapalım.

```python
knn.fit(x_train,y_train)  #knn algoritmasını fit etmemiştik, burada ettik.
print("test accuracy: ",knn.score(x_test,y_test))
```

<font color="Chocolate">
K-fold cross validaton yönteminin k değeri literatürde genellikle 10 seçilir.

## Grid Search with KNN

KNN algoritmasında seçilen k değerinin en uygun skoru verip vermediğini anlayabilmek için Grid Search yapmalıyız.

<br><br><br>
<img src="m6.png" align="left" style="width:800px;height:380px"/>

Yukarıda Grid Search görmekteyiz. Sol taraftaki sayılar KNN algoritmasındaki k değerlerini ifade etmektedir. Gördüğünüz k değerleri için ve k-fold cross validation'un k=3 değeri için veri seti parçalanarak train parçaları ile eğitim ve validation parçası ile test işlemi gerçekleştirilmiştir.

12 farklı parçacık elde edilmiştir. Bir çok parçacık yani grid elde edildiğinden Grid Search Cross Validation denir.

```python
from sklearn.model_selection import GridSearchCV

grid = {"n_neighbors":np.arange(1,50)}
knn = KNeighborsClassifier()

knn_cv = GridSearchCV(knn, grid, cv=10)
knn_cv.fit(x,y)
```

"n_neighbors":np.arange(1,50) ifadesi ile KNN algoritmasındaki k değerinin 1'den 50'ye kadar olan değerleri alması sağlanmıştır.

```python
print("tuned hyperparameter K: ",knn_cv.best_params_) #best_params_ ile en iyi parametre seçilir.
print("tuned parametreye göre en iyi accuracy (best score): ",knn_cv.best_score_)
```

Denenerek bulunan parametre "tuned hyperparameter (ayarlanmış hiperparametre)" olarak adlandırılır.

Çıktıda görüldüğü üzere k=13 olduğunda %98 oranında accuracy değeri elde edilmiştir. En iyi accuracy değeri k=13 olduğunda elde edilmektedir.

## Grid Search with Logistic Regression

Logistic regression için Grid Search uygulayalım.

Logistic regression, binary sınıflandırıcıdır. Output'u binary olan yani mesela 0 ya da 1, kedi ya da köpek gibi veri setlerini tahmin etmekte kullanılan bir sınıflandırıcıdır. Iris veri setinde 3 tane class var ve Logistic regression için 2 class'a düşürelim.

```python
x = x[:100,:]  #İlk 100 değer 2 tane class'ı içerir.
y = y[:100]

from sklearn.linear_model import LogisticRegression

grid = {"C":np.logspace(-3,3,7),"penalty":["l1","l2"]} 
```

"C":np.logspace(-3,3,7) ifadesini açıklayalım: C parametresi, logistic regression regularization parameter diye geçer. Eğer C değeri büyükse overfitting yani aşırı ezberleme durumu olur, eğer C çok küçükse underfitting yani data'yı öğrenememe durumu olur. 

"penalty":["l1","l2"] ifadesini açıklayalım: l1 ve l2 loss fonksiyonlardır. l1:lasso, l2:ridge 'dir.

```python
logreg = LogisticRegression()

logreg_cv = GridSearchCV(logreg, grid, cv=10)
logreg_cv.fit(x,y)
```

```python
print("tuned hyperparameters: (best parameters): ",logreg_cv.best_params_)
```

```python
print("accuracy: ",logreg_cv.best_score_)
```

Tabiki fit etme işleminde x ve y kullanmak pek sağlıklı değil. train_test_split yöntemi ile train ve test setlerini ayırarak fit etme işlemini gerçekleştirmek daha doğru.
