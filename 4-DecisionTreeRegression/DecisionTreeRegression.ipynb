## Decision Tree Regression

Literatürde CART ismiyle geçer. 

CART: Classification and Regression Tree

Regression: Gelecekteki bir değeri tahmin etme işlemidir.

x1 ve x2 isimlerinde feature'ler olduğunu düşünün. Regression modeli sonucunda tahmin etmek istediğimiz değer y olsun. y değeri x1 ve x2 'ye bağımlıdır (dependent). 

Decision tree'nin temel kavramı split'tir. Veri setini "information entropy" kavramına göre split ediyoruz, ayırıyoruz. Bu işlemden sonra bir ağaç elde ederiz. Bu ağaç modelimizdir. Yeni değeri bu ağaca göre predict ederiz.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("decision_tree_regression_dataset.csv",sep=";", header=None)
```

```python
df
```

Veri setinde 0 isimli sütun, tribünde syircinin maçı izlediği seviyedir. Mesela 1 değeri sahaya en yakın olan seviyedir. 1 isimli sütun ise fiyatı ifade ediyor.

Mesela ilk satıra baktığımızda, 1.seviye yerin fiyatı 100 tl olduğunu anlıyoruz.

x ekseni seviyeler, y ekseni de çıktı olacak şekilde kodumuzu yazalım.

```python
x = df.iloc[:,0].values.reshape(-1,1)
y = df.iloc[:,1].values.reshape(-1,1)
```

.values ifadesi ile numpy array'e çevirdiğimizi hatırlayalım.

```python
from sklearn.tree import DecisionTreeRegressor

tree_reg = DecisionTreeRegressor()

tree_reg.fit(x,y)  #Ağaç modelini oluşturduk.

tree_reg.predict([[6]])

y_head = tree_reg.predict(x) #Tüm x değerlerine karşılık gelen y değerini tahmin edip y_head değişkenine atadım.
```

Veri setine baktığımızda x=6 iken y=40 olduğunu görüyoruz. Yukarıdaki kodda x'e 6 değerini verdiğimizde tahmin olarak 40 döndürüyor. Yani doğru tahmin gerçekleştiriyor.

Şimdi görselleştirme yapalım.

```python
plt.scatter(x,y,color="red")
plt.plot(x,y_head,color="green")
plt.xlabel("tribun level")
plt.ylabel("ucret")
plt.show()
```

Daha farklı değerleri x olarak nitelendirip decision tree regression uygulayalım.

```python
x_ = np.arange(min(x),max(x),0.01).reshape(-1,1)  
#minimum x değerinden maksimum x değerine kadar 0.01 aralıklar ile değerler oluştur demektir. 
#Yani 1'den 10'a kadar 0.01 aralıklarla değerler oluşturur.
```

```python
x_
```

```python
y_head = tree_reg.predict(x_)
```

```python
plt.scatter(x,y,color="red")
plt.plot(x_,y_head,color="green")
plt.xlabel("tribun level")
plt.ylabel("ucret")
plt.show()
```
